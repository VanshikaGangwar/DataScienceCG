```python
# Scenario 2: AI Career Advisor & Learning Path Generator
# 🎯 Objective:
# Build a multi-step intelligent advisor system using agent workflows.

# 💼 Problem Statement
# Students need guidance on career paths based on their skills.

# Your system should:

# Analyze student profile
# Suggest career roles
# Generate learning roadmap
# 👥 Agents to Build
# Profile Analyzer Agent
# Understands student background
# Career Recommendation Agent
# Suggests suitable roles
# Skill Gap Agent
# Identifies missing skills
# Learning Path Agent
# Creates step-by-step roadmap

!pip install crewai langchain-community langchain-openai

import os
from crewai import Agent, Task, Crew
from google.colab import userdata

# 3. SET GROQ API KEY

os.environ["GROQ_API_KEY"] = userdata.get("GROQ_API_KEY")


# 4. DEFINE LLM (IMPORTANT FIX)

llm = "groq/llama-3.3-70b-versatile"

# 5. AGENTS

profile_agent = Agent(
    role="Profile Analyzer",
    goal="Analyze student background and extract skills and experience level",
    backstory="Expert in analyzing student profiles",
    llm=llm,
    verbose=True
)

career_agent = Agent(
    role="Career Advisor",
    goal="Suggest suitable career roles based on skills",
    backstory="Expert career counselor",
    llm=llm,
    verbose=True
)

skill_gap_agent = Agent(
    role="Skill Gap Analyzer",
    goal="Identify missing skills required for target roles",
    backstory="Expert in skill analysis",
    llm=llm,
    verbose=True
)

learning_agent = Agent(
    role="Learning Path Generator",
    goal="Create step-by-step roadmap for learning",
    backstory="Expert mentor and educator",
    llm=llm,
    verbose=True
)

# 6. MAIN FUNCTION

def run_system(user_input):

    # Step 1: Analyze Profile
    task1 = Task(
        description=f"""
        Analyze this student profile:
        "{user_input}"

        Extract:
        - Skills
        - Experience level (Beginner/Intermediate/Advanced)
        """,
        expected_output="Structured analysis",
        agent=profile_agent
    )

    # Step 2: Career Recommendation
    task2 = Task(
        description=f"""
        Based on the profile, suggest 2-3 suitable career roles.
        """,
        expected_output="List of roles",
        agent=career_agent,
        context=[task1]
    )

    # Step 3: Skill Gap
    task3 = Task(
        description=f"""
        Identify missing skills required for the suggested roles.
        """,
        expected_output="Skill gaps per role",
        agent=skill_gap_agent,
        context=[task2]
    )

    # Step 4: Learning Path
    task4 = Task(
        description=f"""
        Create a step-by-step learning roadmap for the student.
        """,
        expected_output="Structured roadmap",
        agent=learning_agent,
        context=[task3]
    )


    # 7. CREW

    crew = Crew(
        agents=[profile_agent, career_agent, skill_gap_agent, learning_agent],
        tasks=[task1, task2, task3, task4],
        verbose=True
    )

    result = crew.kickoff()
    return result


# =============================
# 8. USER INPUT LOOP
# =============================

print("🎓 AI Career Advisor (type 'exit' to quit)\n")

while True:
    user_input = input("\n Enter your profile (or type 'exit'): ")

    if user_input.strip().lower() == "exit":
        print(" Exiting system...")
        break

    if user_input.strip() == "":
        print("Please enter something!")
        continue

    try:
        output = run_system(user_input)

        print("\n FINAL RESULT:\n")
        print(output)
        print("\n" + "="*50)

    except Exception as e:
        print(" Error:", e)

```
