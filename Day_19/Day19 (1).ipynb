```python

# Smart Shopping Assistant 🛒


# Objective:
# Build an agent that finds the best chocolate cake under ₹500
# using step-by-step reasoning and modular tools
# The Smart Shopping Assistant transforms a simple user query into a
#  structured decision-making process by integrating data retrieval, constraint filtering,
#   and optimal selection, thereby demonstrating the core principles of Agentic AI



# Step 1 — Define Tools



# Smart Shopping Assistant 🛒 (Agentic AI)


# Objective:
# Build an intelligent agent that finds the best chocolate cake under ₹500
# using planning, tool execution, and reasoning



# Step 1 — Define Tools


def search_cakes(query):
    return [
        {"shop": "Sweet Treats", "cake": "Chocolate Cake", "price": 450, "rating": 4.5},
        {"shop": "Cake World", "cake": "Chocolate Truffle", "price": 550, "rating": 4.7},
        {"shop": "Bake Bliss", "cake": "Dark Chocolate Cake", "price": 480, "rating": 4.6},
        {"shop": "Cake Corner", "cake": "Chocolate Fudge Cake", "price": 400, "rating": 4.4}
    ]


def filter_by_budget(cakes, budget):
    return [cake for cake in cakes if cake["price"] <= budget]


def select_best_cake(cakes):
    if not cakes:
        return None
    return max(cakes, key=lambda x: x["rating"])



# Step 2 — Agent Brain (Planning)


def agent_brain(user_query):
    plan = []

    if "cake" in user_query.lower():
        plan.append("search_cakes")
        plan.append("filter_budget")
        plan.append("choose_best")

    return plan



# Step 3 — Execution Engine


def execute_plan(plan):
    context = {}

    for step in plan:

        if step == "search_cakes":
            print("\n[Action] Searching Cakes...")
            context["cakes"] = search_cakes("chocolate cake")
            print(context["cakes"])

        elif step == "filter_budget":
            print("\n[Action] Filtering by Budget...")
            context["filtered"] = filter_by_budget(context["cakes"], 500)
            print(context["filtered"])

        elif step == "choose_best":
            print("\n[Action] Selecting Best Cake...")
            best = select_best_cake(context["filtered"])
            context["best_cake"] = best
            print(best)

    return context



# Step 4 — Reasoning Layer


def choose_best_cake_reasoning(cakes):
    if not cakes:
        return "No cake available within budget."

    best = max(cakes, key=lambda x: x["rating"])
    return f"Choose {best['cake']} from {best['shop']} at ₹{best['price']} (Rating: {best['rating']})"



# Step 5 — Full Agent


def shopping_agent(user_query):
    print("User Query:", user_query)

    # Step 1: Plan
    plan = agent_brain(user_query)
    print("\n[Plan]", plan)

    # Step 2: Execute
    result = execute_plan(plan)

    # Step 3: Final Answer
    final_output = f"""
 Best Cake: {result['best_cake']['cake']} from {result['best_cake']['shop']}
 Price: ₹{result['best_cake']['price']}
 Rating: {result['best_cake']['rating']}
"""

    # Step 4: Reasoning
    reasoning = choose_best_cake_reasoning(result["filtered"])

    return final_output + "\nReasoning: " + reasoning



# Step 6 — Run Agent


response = shopping_agent("Find best chocolate cake under ₹500")
print("\nFinal Answer:\n", response)
```

```python
# Scenario: The University’s AI Plagiarism Agent

# A university deploys an intelligent system called PlagiarismLLM to automatically evaluate student essays for originality. Instead of relying on manual checking, the system operates as an agentic AI, meaning it performs reasoning, planning, and action before giving a final decision.

# When a student submits an essay, the system first analyzes the content to detect patterns such as repeated phrases, similarity with known sources, or unusually consistent writing styles. Based on this analysis, it forms an internal understanding of whether the essay might contain copied content.

# Next, the system decides what action to take. If strong indicators of plagiarism are found, it plans to flag the essay. Otherwise, it proceeds to mark the work as original.

# The system then executes the chosen action using predefined tools. These tools return a clear classification result.

# Finally, the agent presents a decision that can be used by faculty for further evaluation. This entire process demonstrates a structured, goal-driven AI system rather than a simple rule-based checker.


# Scenario: The University’s AI Plagiarism Agent
# Imagine a university has deployed an AI-powered plagiarism detection agent named PlagiarismLLM.
# Its job is to monitor student essay submissions in real time and decide whether they are original or potentially plagiarized.

# Step 1: Thinking (Reasoning)
# Whenever an essay is submitted, PlagiarismLLM first analyzes the text:
# - Example: “Essay contains large verbatim passages from online sources”
# - PlagiarismLLM thinks: “Analyzing essay: multiple verbatim passages detected.”

# Step 2: Planning (Choosing an Action)
# Based on the thought, PlagiarismLLM decides which tool to use:
# - If the essay has long verbatim matches or suspicious similarity scores, it plans to flag as potential plagiarism.
# - Otherwise, it plans to mark as original work.

# Step 3: Acting (Running Tools)
# PlagiarismLLM hands over the decision to PlagiarismTools, which executes the action:
# - flag_plagiarism → returns “Potential Plagiarism Detected”
# - mark_original → returns “Original Work”

# Step 4: Answering (Final Decision)
# Once a clear classification is reached, PlagiarismLLM announces the decision:
# - “Decision: Potential Plagiarism Detected”
# - “Decision: Original Work”


# AI Plagiarism Detection Agent


# Objective:
# Build an agent that evaluates essays and classifies them as
# "Potential Plagiarism Detected" or "Original Work"
# using reasoning, planning, and execution


# Step 1 — Define Tools

def analyze_essay(essay_text):
    """Analyzes essay content and returns reasoning signals"""
    if "copy" in essay_text.lower() or "paste" in essay_text.lower():
        return {"verdict": "suspicious", "reason": "Verbatim patterns detected"}
    if len(essay_text.split()) < 50:
        return {"verdict": "suspicious", "reason": "Essay too short"}
    return {"verdict": "clean", "reason": "No major similarity detected"}


def flag_plagiarism():
    """Tool to flag plagiarism"""
    return "Potential Plagiarism Detected"


def mark_original():
    """Tool to mark essay as original"""
    return "Original Work"


# Step 2 — Agent Brain (Reasoning + Planning)

def agent_brain(analysis):
    """Decides which action to take based on analysis"""
    plan = []

    if analysis["verdict"] == "suspicious":
        plan.append("flag")
    else:
        plan.append("approve")

    return plan


# Step 3 — Execution Engine

def execute_plan(plan):
    """Executes the selected tool"""
    result = None

    for step in plan:
        if step == "flag":
            print("[Action] Flagging plagiarism...")
            result = flag_plagiarism()

        elif step == "approve":
            print("[Action] Marking as original...")
            result = mark_original()

    return result


# Step 4 — Full Agent Pipeline

def plagiarism_agent(essay_text):
    print("Essay Received\n")

    print("[Thinking] Analyzing essay...")
    analysis = analyze_essay(essay_text)
    print("Analysis:", analysis)

    print("\n[Planning] Deciding action...")
    plan = agent_brain(analysis)
    print("Plan:", plan)

    print("\n[Acting] Executing decision...")
    decision = execute_plan(plan)

    print("\n[Answer]")
    final_output = f"Decision: {decision}"
    return final_output


# Step 5 — Run Example

essay_input = "This essay contains copy paste content from internet sources."
response = plagiarism_agent(essay_input)

print("\nFinal Output:\n", response)
```

```python
from openai import OpenAI
from typing import Dict, Any, Optional
from datetime import datetime, timedelta
import re
import json
import random

# Initialize Groq client
client = OpenAI(
    base_url="https://openai/v1",
    api_key="id1Njv1R6r"
)

class AppointmentTools:
    """Tool executor for appointment actions"""

    def __init__(self):
        # Mock appointment database
        self.appointments_db = {
            "dr_sharma": {
                "2026-03-19": ["09:00", "10:00", "11:00", "14:00", "15:00", "16:00"],
                "2026-03-20": ["09:00", "10:00", "11:00", "14:00", "15:00", "16:00"],
                "2026-03-21": ["09:00", "11:00", "14:00", "15:00", "16:00"]  # 10 AM booked
            }
        }

    def check_availability(self, doctor: str, date: str, time: str) -> bool:
        """Check if slot is available"""
        doctor_key = doctor.lower().replace(" ", "_").replace("dr.", "dr")

        if doctor_key in self.appointments_db:
            if date in self.appointments_db[doctor_key]:
                return time in self.appointments_db[doctor_key][date]
        return False

    def confirm_slot(self, doctor: str, date: str, time: str, patient: str = "Patient") -> str:
        """Tool 1: Confirm appointment slot"""
        confirmation_id = f"APT{random.randint(10000, 99999)}"

        return f"""
╔════════════════════════════════════════════════════════════════╗
║                    ✅ APPOINTMENT CONFIRMED                     ║
╠════════════════════════════════════════════════════════════════╣
║  Confirmation ID: {confirmation_id}
║  Patient: {patient}
║  Doctor: Dr. {doctor}
║  Date: {date}
║  Time: {time}
║  Status: ✅ Confirmed
║  Instructions: Please arrive 15 minutes early
╚════════════════════════════════════════════════════════════════╝
"""

    def suggest_reschedule(self, doctor: str, date: str, time: str, patient: str = "Patient") -> str:
        """Tool 2: Suggest rescheduling options"""
        # Find available slots
        doctor_key = doctor.lower().replace(" ", "_").replace("dr.", "dr")
        available_slots = []

        if doctor_key in self.appointments_db and date in self.appointments_db[doctor_key]:
            all_slots = ["09:00", "10:00", "11:00", "14:00", "15:00", "16:00"]
            booked = self.appointments_db[doctor_key][date]
            available_slots = [slot for slot in all_slots if slot not in booked]

        alternatives = ", ".join(available_slots[:3]) if available_slots else "tomorrow"

        return f"""
╔════════════════════════════════════════════════════════════════╗
║                    ⚠️ SLOT UNAVAILABLE                          ║
╠════════════════════════════════════════════════════════════════╣
║  Patient: {patient}
║  Doctor: Dr. {doctor}
║  Date: {date}
║  Requested Time: {time} - ❌ NOT AVAILABLE
║
║  📅 Available alternatives today:
║  {alternatives}
║
║  Please contact reception to reschedule: 555-MEDICAL
╚════════════════════════════════════════════════════════════════╝
"""

    def suggest_alternative_doctor(self, doctor: str, date: str, time: str, patient: str = "Patient") -> str:
        """Tool 3: Suggest alternative doctor"""
        return f"""
╔════════════════════════════════════════════════════════════════╗
║                    👨‍⚕️ ALTERNATIVE DOCTOR                        ║
╠════════════════════════════════════════════════════════════════╣
║  Dr. {doctor} is fully booked on {date}
║
║  Available doctors:
║  • Dr. Patel - {time} available
║  • Dr. Singh - {time} available
║  • Dr. Kumar - {time} available
║
║  Would you like to book with an alternative?
╚════════════════════════════════════════════════════════════════╝
"""

class MediLLM:
    """Hospital AI Appointment Agent"""

    def __init__(self):
        self.tools = AppointmentTools()
        self.patient_context = {}

    def parse_request(self, request: str) -> Dict[str, str]:
        """Extract doctor, date, time from request"""
        # Simple parsing logic
        request_lower = request.lower()

        # Extract doctor
        doctor_match = re.search(r'dr\.?\s*(\w+)', request_lower)
        doctor = doctor_match.group(1).title() if doctor_match else "Sharma"

        # Extract time
        time_match = re.search(r'(\d{1,2})\s*(?::| )?(\d{2})?\s*(am|pm)?', request_lower)
        if time_match:
            hour = time_match.group(1)
            minute = time_match.group(2) if time_match.group(2) else "00"
            meridiem = time_match.group(3) if time_match.group(3) else "am"
            time = f"{hour}:{minute} {meridiem.upper()}"
        else:
            time = "10:00 AM"

        # Extract date
        if "tomorrow" in request_lower:
            date = (datetime.now() + timedelta(days=1)).strftime("%Y-%m-%d")
        elif "today" in request_lower:
            date = datetime.now().strftime("%Y-%m-%d")
        else:
            date = (datetime.now() + timedelta(days=1)).strftime("%Y-%m-%d")  # Default tomorrow

        # Extract patient name
        patient_match = re.search(r'for\s+([a-z\s]+?)(?:\s+at|\s+with|\s+on|$)', request_lower)
        patient = patient_match.group(1).strip().title() if patient_match else "Patient"

        return {
            "doctor": doctor,
            "date": date,
            "time": time,
            "patient": patient,
            "raw_request": request
        }

    def step1_think(self, parsed_request: Dict[str, str]) -> str:
        """Step 1: Reasoning - Analyze the request"""
        thought = f"""
🧠 STEP 1: THINKING (Analyzing Request)
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
🤔 MediLLM analyzing:
   Patient: {parsed_request['patient']}
   Doctor: Dr. {parsed_request['doctor']}
   Date: {parsed_request['date']}
   Time: {parsed_request['time']}

📋 Checking availability in hospital system...
"""
        print(thought)
        return thought

    def step2_plan(self, parsed_request: Dict[str, str]) -> str:
        """Step 2: Planning - Choose appropriate action"""

        # Check availability
        is_available = self.tools.check_availability(
            parsed_request['doctor'],
            parsed_request['date'],
            parsed_request['time'].split()[0]  # Extract just the time part
        )

        if is_available:
            plan = f"""
📋 STEP 2: PLANNING (Choosing Action)
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
✅ Slot is AVAILABLE!
   Action Selected: confirm_slot
   Tool: AppointmentTools.confirm_slot()
   Parameters: doctor={parsed_request['doctor']}, date={parsed_request['date']},
               time={parsed_request['time']}, patient={parsed_request['patient']}
"""
            self.selected_action = "confirm"
        else:
            # Check if alternative same day slots exist
            has_alternatives = random.choice([True, False])  # Simplified logic

            if has_alternatives:
                plan = f"""
📋 STEP 2: PLANNING (Choosing Action)
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
❌ Slot {parsed_request['time']} is BOOKED!
   ✅ Alternative slots available today
   Action Selected: suggest_reschedule
   Tool: AppointmentTools.suggest_reschedule()
   Parameters: doctor={parsed_request['doctor']}, date={parsed_request['date']},
               time={parsed_request['time']}, patient={parsed_request['patient']}
"""
                self.selected_action = "reschedule"
            else:
                plan = f"""
📋 STEP 2: PLANNING (Choosing Action)
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
❌ Dr. {parsed_request['doctor']} fully booked on {parsed_request['date']}
   ✅ Alternative doctors available
   Action Selected: suggest_alternative_doctor
   Tool: AppointmentTools.suggest_alternative_doctor()
   Parameters: doctor={parsed_request['doctor']}, date={parsed_request['date']},
               time={parsed_request['time']}, patient={parsed_request['patient']}
"""
                self.selected_action = "alternative"

        print(plan)
        return plan

    def step3_act(self, parsed_request: Dict[str, str]) -> str:
        """Step 3: Acting - Execute the chosen tool"""

        print("\n⚙️ STEP 3: ACTING (Running Tools)")
        print("━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━")

        if self.selected_action == "confirm":
            result = self.tools.confirm_slot(
                parsed_request['doctor'],
                parsed_request['date'],
                parsed_request['time'].split()[0],
                parsed_request['patient']
            )
        elif self.selected_action == "reschedule":
            result = self.tools.suggest_reschedule(
                parsed_request['doctor'],
                parsed_request['date'],
                parsed_request['time'].split()[0],
                parsed_request['patient']
            )
        else:
            result = self.tools.suggest_alternative_doctor(
                parsed_request['doctor'],
                parsed_request['date'],
                parsed_request['time'].split()[0],
                parsed_request['patient']
            )

        print(result)
        return result

    def step4_answer(self, action_result: str) -> str:
        """Step 4: Answer - Final decision announcement"""

        print("\n🎯 STEP 4: ANSWER (Final Decision)")
        print("━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━")

        if "CONFIRMED" in action_result:
            decision = "Decision: ✅ APPOINTMENT CONFIRMED"
        elif "UNAVAILABLE" in action_result:
            decision = "Decision: ⚠️ SLOT UNAVAILABLE - PLEASE RESCHEDULE"
        else:
            decision = "Decision: 👨‍⚕️ ALTERNATIVE DOCTOR SUGGESTED"

        final = f"""
🏥 MEDILLM FINAL DECISION:
{decision}

{action_result}
"""
        print(final)
        return final

    def process_request(self, request: str) -> str:
        """Main pipeline to process patient request"""

        print("\n" + "=" * 70)
        print("🏥 HOSPITAL AI APPOINTMENT AGENT - MEDILLM".center(70))
        print("=" * 70)

        # Parse the request
        parsed = self.parse_request(request)

        # Step 1: Think
        self.step1_think(parsed)

        # Step 2: Plan
        self.step2_plan(parsed)

        # Step 3: Act
        action_result = self.step3_act(parsed)

        # Step 4: Answer
        final_decision = self.step4_answer(action_result)

        return final_decision

# Patient request simulator
def patient_request_simulator():
    """Simulate various patient requests"""

    medi = MediLLM()

    test_requests = [
        "Request for Dr. Sharma at 10 AM tomorrow for John",
        "Need appointment with Dr. Patel at 2 PM today for Sarah",
        "Book Dr. Sharma at 11 AM tomorrow for Michael",
        "Emergency appointment with Dr. Kumar at 3 PM for baby Emma",
        "Follow-up with Dr. Sharma at 10 AM today for Robert"
    ]

    for i, request in enumerate(test_requests, 1):
        print(f"\n{'#' * 70}")
        print(f"PATIENT REQUEST #{i}".center(70))
        print(f"Request: '{request}'")
        print(f"{'#' * 70}")

        medi.process_request(request)

        if i < len(test_requests):
            input("\nPress Enter for next patient request...")

# Quick test
if __name__ == "__main__":
    # Single request
    medi = MediLLM()
    result = medi.process_request("Request for Dr. Sharma at 10 AM tomorrow for Amit Kumar")
    print(result)

    # Uncomment for multiple test cases
    # patient_request_simulator()
```
