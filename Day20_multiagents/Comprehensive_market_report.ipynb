```python
#Install & Setup
from google.colab import userdata
from groq import Groq

api_key = userdata.get("GROQ_API_KEY")

client = Groq(api_key=api_key)

#LLM Helper
def ask_groq(prompt):
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content


#Shared Memory (Message Bus)
class SharedMemory:
    def __init__(self):
        self.data = {}

    def update(self, agent_name, result):
        self.data[agent_name] = result

    def get(self, key=None):
        if key:
            return self.data.get(key, "")
        return self.data

#Search Agent
class SearchAgent:
    name = "search"

    def run(self, goal, memory):
        print("[Search Agent] Searching data...")

        prompt = f"""
        Find key trends, statistics, and insights about:
        {goal}
        Provide structured bullet points.
        """

        result = ask_groq(prompt)
        return result

#Analyst Agent
class AnalystAgent:
    name = "analyst"

    def run(self, goal, memory):
        print("[Analyst Agent] Analyzing data...")

        search_data = memory.get("search")

        prompt = f"""
        Analyze the following data and extract insights:
        {search_data}

        Provide key insights, patterns, and conclusions.
        """

        return ask_groq(prompt)

#Writer Agent
class WriterAgent:
    name = "writer"

    def run(self, goal, memory):
        print("[Writer Agent] Writing report...")

        analysis = memory.get("analyst")

        prompt = f"""
        Write a detailed market report based on:
        {analysis}

        Include:
        - Introduction
        - Market trends
        - Future outlook
        """

        return ask_groq(prompt)

#QA Agent
class QAAgent:
    name = "qa"

    def run(self, goal, memory):
        print("[QA Agent] Reviewing report...")

        report = memory.get("writer")

        prompt = f"""
        Review the following report:
        {report}

        Improve clarity, correctness, and tone.
        Return final polished version.
        """

        return ask_groq(prompt)

#Orchestrator
class Orchestrator:

    def __init__(self):
        self.agents = [
            SearchAgent(),
            AnalystAgent(),
            WriterAgent(),
            QAAgent()
        ]
        self.memory = SharedMemory()

    def run(self, goal):
        print("\n🚀 Starting AI Research Pipeline...\n")

        for agent in self.agents:
            result = agent.run(goal, self.memory)
            self.memory.update(agent.name, result)

        return self.memory.get("qa")

goal = "Write a comprehensive market report on EV industry trends in India for 2025"

system = Orchestrator()
final_report = system.run(goal)

print("\n📄 FINAL REPORT:\n")
print(final_report)


```
