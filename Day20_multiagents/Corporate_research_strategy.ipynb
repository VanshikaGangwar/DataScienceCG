```python
# Scenario: Corporate Research Strategy
# Multiple expert agents collaborate to analyze a business goal:
# - Market Research Expert: Analyzes demand, competition, customer preferences.
# - Finance Expert: Analyzes budget, ROI, pricing strategy.
# - Operations Expert: Analyzes supply chain, production capacity, logistics.
# - Legal Expert: Analyzes regulations, compliance, IP issues.
# - Strategy Head: Combines all insights into a final recommendation.
```

```python
import os
import requests
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("NVIDIA_API_KEY")

def call_llm(prompt):
    url = "https://integrate.api.nvidia.com/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    data = {
        "model": "meta/llama3-70b-instruct",
        "messages": [{"role": "user", "content": prompt}],
        "temperature": 0.3
    }

    res = requests.post(url, headers=headers, json=data)
    return res.json()["choices"][0]["message"]["content"]
```

```python
def market_research_worker(goal):
    prompt = f"""
    You are a Market Research Expert.

    Goal: {goal}

    Analyze:
    - Demand
    - Competition
    - Customer preferences

    Give short insights.
    """
    return call_llm(prompt)


def finance_worker(goal):
    prompt = f"""
    You are a Finance Expert.

    Goal: {goal}

    Analyze:
    - Budget required
    - ROI
    - Pricing strategy

    Give short insights.
    """
    return call_llm(prompt)


def operations_worker(goal):
    prompt = f"""
    You are an Operations Expert.

    Goal: {goal}

    Analyze:
    - Supply chain
    - Production capacity
    - Logistics

    Give short insights.
    """
    return call_llm(prompt)


def legal_worker(goal):
    prompt = f"""
    You are a Legal Expert.

    Goal: {goal}

    Analyze:
    - Regulations
    - Compliance
    - IP issues

    Give short insights.
    """
    return call_llm(prompt)
```

```python
def strategy_head(goal, insights):
    prompt = f"""
    You are Strategy Head.

    Business Goal: {goal}

    Expert Insights:
    {insights}

    Provide a final strategic recommendation in 4-5 lines.
    """
    return call_llm(prompt)
```

```python
def run_research_system(goal):
    print(f"Business Goal: {goal}\n")

    insights = {
        "Market Research": market_research_worker(goal),
        "Finance": finance_worker(goal),
        "Operations": operations_worker(goal),
        "Legal": legal_worker(goal)
    }

    final = strategy_head(goal, insights)
    print("\nFINAL STRATEGY:\n")
    print(final)


# TEST
run_research_system("Launch a new electric vehicle in the Indian market")
```
